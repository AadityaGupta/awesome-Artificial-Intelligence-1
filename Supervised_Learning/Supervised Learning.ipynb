# What is Supervised Learning ?


Supervised learning is where you have input variables (x) and an output variable (Y) and you use an algorithm to learn the mapping function from the input to the output.

Y = f(X)

The goal is to approximate the mapping function so well that when you have new input data (x) that you can predict the output variables (Y) for that data.

It is called supervised learning because the algorithm is learned from the training dataset.

Example - Suppose faculty wants each and every student of his class to score more than 80% in exam. He knows what should be th output(>80% marks) and input(syllabus he's going to teach). Now how well he teaches comes into a picture of LEARNING MECHANISM.

## Types of Supervised Learning :

Supervised learning problems in turn are categorized into :
- Classification problem. 
- Regression Problem 





## What is Classification Problem?

The main goal of classification is to predict the target class (Yes/ No). If the trained  model is for predicting any of **two target classes**, it is known as **binary classification**. Examples :
- Considering the student profile to predict whether the student will pass or fail. 
- Considering the customer, transaction details to predict whether he will buy the new product or not.
These kind problems will be addressed with binary classification. <br>

If we have to predict more the two target classes, then it is known as **multi-classification**. 
- Considering all subject details of a student to  predict which subject the student will score more. 
- Identifying the object in an image.

## What is Regression Problem?

A regression problem is the one where the input is mapped to a continues set of output values. In other words, a regression problem is a problem in which the output variable is a real value, such as “dollars” or “weight”.

**Best Example : House Price Prediction**
Let's suppose you want to sell a house and you don’t know the price which you can take — it can’t be too low or too high. To find house price you usually try to find similar properties in your neighborhood and based on gathered data you will try to assess your house price

If you clearly observe the problem is about finding the value. Keep in mind, here we are forecasting a value ( Prediction ) and the other thing this prediction also belongs to Supervised learning. This is because you are learning this from you train data.
